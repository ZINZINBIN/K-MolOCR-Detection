```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
import os
from PIL import Image

table_images = os.listdir('./dataset/cropped_table_images')
print(len(table_images))
imgs = []
for i in range(20):
    img_PIL = Image.open('./dataset/cropped_table_images/'+table_images[i])
    img = np.array(img_PIL)
    imgs.append(img)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
import shutil
for i in range(10):
    table_images = os.listdir('./dataset/cropped_table_images')
    src = './dataset/cropped_table_images/'+table_images[i]
    dst = './dataset/sample_table/image_{:05d}.png'.format(i+1)
    shutil.copyfile(src, dst)
```
